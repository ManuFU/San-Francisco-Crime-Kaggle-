```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# load the data from CSV file
sc = pd.read_csv('train.csv')
```

```python
#settings
dates = sc['Dates']
category = sc['Category']
descript = sc['Descript']
dow = sc['DayOfWeek']
pd = sc['PdDistrict']
res= sc['Resolution']
address = sc['Address']
X = sc['X']
Y = sc['Y']
```

```python
#Crime per District
sc_CrimePerDistrict = sc.drop(['Dates', 'Descript', 'DayOfWeek','Resolution', 'Address', 'X', 'Y' ], axis=1)
#sc_CrimePerDistrict.set_index(pd, inplace = True)
sc_CrimePerDistrict

#Pivot
CrimePerDistrict = sc_CrimePerDistrict.pivot_table(sc_CrimePerDistrict, index = category, columns = pd, aggfunc='count')
CrimePerDistrict.head()
#Plotting
CrimePerDistrict.plot(kind="bar", figsize=(21, 21))
#Saving
plt.savefig('CrimePerWeekday.png')
plt.show()
```
