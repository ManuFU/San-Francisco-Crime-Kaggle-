```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pylab

#size of figure
pylab.rcParams['figure.figsize'] = (12, 15)

filename = 'train.csv'
#load file 
data = pd.read_csv(filename, parse_dates=['Dates'], index_col='Dates')
#count values per Category
crimes_rating = data['Category'].value_counts()
y_pos = np.arange(len(crimes_rating[0:18].keys()))

#create bar-plot
plt.barh(y_pos, crimes_rating[0:18].get_values())

#label it
plt.xlabel('Anzahl der Straftaten', fontsize = 18)
plt.ylabel('Straftat', fontsize = 18) 

#plot and save it
plt.yticks(y_pos, map(lambda x:x.title(),crimes_rating[0:18].keys()), fontsize = 14)
plt.title('Summe pro Verbrechen', fontsize = 28)
plt.savefig('Sum_Crimes.png')
plt.show()
```
