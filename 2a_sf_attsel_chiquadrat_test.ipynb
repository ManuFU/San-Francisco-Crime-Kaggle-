```python
# Import Libraries 
import os
import pandas as pd
import numpy as np
import researchpy as rp

from scipy import stats
```

```python
#Lese CSV ein und caste Dates zu datetime64
def open_train():
    dataset = pd.read_csv('train.csv')
    dataset['Dates'] = pd.to_datetime(dataset['Dates']) 
    return dataset

def open_test():
    dataset = pd.read_csv('test.csv')
    dataset['Dates'] = pd.to_datetime(dataset['Dates']) 
    return dataset
```

```python
#Chi-Quadrat Feature Selection
def chi_square(chi_data):
    results_catyears = rp.crosstab(chi_data['Category'], chi_data['years'], prop= 'col', test= 'chi-square')
    results_catmonths = rp.crosstab(chi_data['Category'], chi_data['months'], prop= 'col', test= 'chi-square')
    results_catdays = rp.crosstab(chi_data['Category'], chi_data['days'], prop= 'col', test= 'chi-square')
    results_cathours = rp.crosstab(chi_data['Category'], chi_data['hours'], prop= 'col', test= 'chi-square')
    results_catminutes = rp.crosstab(chi_data['Category'], chi_data['minutes'], prop= 'col', test= 'chi-square')
    results_catseconds = rp.crosstab(chi_data['Category'], chi_data['seconds'], prop= 'col', test= 'chi-square')
    results_catdis = rp.crosstab(chi_data['Category'], chi_data['Descript'], prop= 'col', test= 'chi-square')
    results_catday = rp.crosstab(chi_data['Category'], chi_data['DayOfWeek'], prop= 'col', test= 'chi-square')
    results_catpd = rp.crosstab(chi_data['Category'], chi_data['PdDistrict'], prop= 'col', test= 'chi-square')
    results_catres = rp.crosstab(chi_data['Category'], chi_data['Resolution'], prop= 'col', test= 'chi-square')
    results_catadd = rp.crosstab(chi_data['Category'], chi_data['Address'], prop= 'col', test= 'chi-square')
    results_catY = rp.crosstab(chi_data['Category'], chi_data['Y'], prop= 'col', test= 'chi-square')
    results_catX = rp.crosstab(chi_data['Category'], chi_data['X'], prop= 'col', test= 'chi-square')
    return print(results_catyears, results_catmonths, results_catdays, results_cathours, results_catminutes, 
                 results_catseconds, results_catdis, results_catday, results_catpd, results_catres,
                 results_catadd, results_catY, results_catX) 
```

```python

train_data = open_train()
test_data = open_test() 

#Uppercase
train_data['DayOfWeek'] = train_data['DayOfWeek'].str.upper()
train_data['Address'] = train_data['Address'].str.upper()
test_data['DayOfWeek'] = train_data['DayOfWeek'].str.upper()
test_data['Address'] = train_data['Address'].str.upper()

#Show amount of Data
print('Columns Trainset: ')
print(train_data.shape)
print('Columns Testset: ')
print(test_data.shape)

#Drop Duplicates
train_data = train_data.drop_duplicates()
print('Columns after Duplicates Trainset: ')
print(train_data.shape)

#Outliers Elimination
train_data = train_data[(np.abs(stats.zscore(train_data['X'])) < 3)]
train_data = train_data[(np.abs(stats.zscore(train_data['Y'])) < 3)]
print('Columns after after Z-Score Analysis Trainset: ')
print(train_data.shape)

#Split Data
for x in [train_data, test_data]:
    x['years'] = x['Dates'].dt.year
    x['months'] = x['Dates'].dt.month
    x['days'] = x['Dates'].dt.day
    x['hours'] = x['Dates'].dt.hour
    x['minutes'] = x['Dates'].dt.minute
    x['seconds'] = x['Dates'].dt.second

#Drop Dates
train_data = train_data.drop(['Dates'], axis = 1)
test_data = test_data.drop(['Dates'], axis = 1)
    
#Chi-square Test
chi_data = train_data
chi_square(chi_data)
    
```
